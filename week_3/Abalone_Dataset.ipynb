```python
#Question 3
def Summary(x):
    x=x.sort_values()
    mini = x[0]
    q1 = x[int((len(x)+1)/4)]
    q2 = x[int((2*len(x)+1)/4)]
    q3 = x[int((3*len(x) +1)/4)]
    maxi = x[len(x)-1]
    
    print("Min :" , mini)
    print("Q1 :" , q1)
    print("Q2 :" , q2)
    print("Q3 :" , q3)
    print("Max :" , maxi)
    
print("Length :")
Summary(data.Length)
print()
print("Diameter :")
Summary(data.Diameter)
print()
print("Height :")
Summary(data.Height)
print()
print("Whole weight :")
Summary(data['Whole weight'])
print()
print("Shucked weight :")
Summary(data['Shucked weight'])
print()
print("viscera weight :")
Summary(data['Viscera weight'])
print()
print("Shell weight :")
Summary(data['Shell weight'])
print()
print("Rings :")
Summary(data.Rings)
```
