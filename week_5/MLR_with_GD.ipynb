```python
#splitting the dataset into x and y
y = np.array(df['MEDV'])
x = df[['RM','LSTAT','PTRATIO']]
```

```python
#performing multilinear regression using the gd algorithm
import math
def multiple_linear_regression(X, y, learning_rate=0.01, num_iterations=1000):
    # Initialize weights and bias to zero
    n_samples, n_features = X.shape
    weights = np.zeros(n_features)
    bias = 0
    
    # Perform gradient descent
    for i in range(num_iterations):
        y_pred = np.dot(X, weights) + bias

        cost = (1/n_samples)*np.sum((y-y_pred))

        # Calculate derivatives of weights and bias
        dw = -(2/n_samples) * np.dot(X.T, (y-y_pred))
        db = -(2/n_samples) * np.sum(y-y_pred)

        # Update weights and bias
        weights -= learning_rate * dw
        bias -= learning_rate * db
        print("m {},c {},cost {},iteration {}".format(weights,bias,cost,i+1))

  
multiple_linear_regression(x,y,0.001,10000)
```
