```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('titanic.csv')
data.head()
```

```python
data.info()
```

```python
data_final = data[[ 'Pclass', 'Sex', 'Age', 'SibSp','Parch', 'Embarked','Survived']]
data_final.head()
```

```python
data_final.info()
```

```python
#preproccessing
print(data_final.isnull().sum())
mean_age = round(data_final['Age'].mean(),2)
print(mean_age)
data_final['Age'] = data_final['Age'].fillna(mean_age)
print(data_final.isnull().sum())
data_final = data_final.dropna()
print(data_final.isnull().sum())
len(data_final)
data.head()
```

```python
filter1 = data_final['Age']<70
filter2 = data_final['Age']>1
data_new = data_final[filter1 & filter2]
data_new.head()
```

```python
df = data_new.drop(["Sex","Embarked"],axis=1)
```

```python
y = df['Survived']
x = df.drop(["Survived"],axis=1)

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size=0.85,random_state=0)
df.head()
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(xtrain,ytrain)
```

```python
ypred = model.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest,ypred)
cm
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
print(accuracy_score(ytest,ypred))
print(recall_score(ytest,ypred))
print(precision_score(ytest,ypred))
print(f1_score(ytest,ypred))
```

```python
#performing logistic regression using gradient descent
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def logistic_regression(X, y, learning_rate=0.01, num_iterations=1000):
    # Initialize weights and bias to zero
    n_samples, n_features = X.shape
    weights = np.zeros(n_features)
    bias = 0
    
    # Perform gradient descent
    for i in range(num_iterations):
        z = np.dot(X, weights) + bias
        y_pred = sigmoid(z)
        cost = -(1/n_samples) * np.sum(y*np.log(y_pred) + (1-y)*np.log(1-y_pred))
        # Calculate derivatives of weights and bias
        dw = -(2/n_samples) * np.dot(X.T, (y-y_pred))
        db = -(2/n_samples) * np.sum(y-y_pred)

        # Update weights and bias
        weights -= learning_rate * dw
        bias -= learning_rate * db

        print("m {},c {},cost {},iteration {} ".format(weights,bias,cost,i+1))

y = df['Survived']
x = df.drop(["Survived"],axis=1)

logistic_regression(x,y,0.002,100)
        
```
